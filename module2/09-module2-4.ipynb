# Gestion durable
## Provenance de l'eau potable pour Montréal
Pour **protéger les sources** d'approvisionnement en eau potable, le **ministère** de l’Environnement, de la Lutte contre les changements climatiques, de la Faune et des Parcs (MELCCFP) du Québec a décidé d’**interdire certaines activités**. Par exemple, à proximité des prises d’eau, il interdit le pâturage, soit les espaces couverts d'herbes réservés à la nourriture du bétail; l'épandage de matières fertilisantes azotées; et le stockage de déjections animales à même le sol. Dans son [Règlement sur le prélèvement des eaux et leur protection (RPEP)](https://www.legisquebec.gouv.qc.ca/fr/document/rc/Q-2,%20r.%2035.2), il a donc défini des périmètres à respecter : les aires de protection immédiates, intermédiaires et éloignées. Chaque usine a ses **aires de protection**, établies selon une distance précise. Nous verrons qu'il est également nécessaire de faire l'**analyse de la vulnérabilité** de la source d’eau de chacune des installations de production d’eau potable. Mais avant tout, penchons-nous sur les endroits où les six {term}`usines de production <Usine de production>` d'eau potable de Montréal puisent leur eau.
