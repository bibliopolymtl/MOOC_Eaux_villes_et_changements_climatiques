## Eutrophisation : activités humaines et changements climatiques
### Qu'est-ce que l’eutrophisation

L’eutrophisation est un processus par lequel un plan d’eau passe d’un niveau {term}`oligotrophe` (pauvre) à {term}`eutrophe` (riche), c’est-à-dire qu’il atteint une **concentration de nutriments** très élevée et une grande **accumulation des {term}`sédiments <Sédiment>`**. Ses caractéristiques sont alors modifiées, puis la croissance des **plantes aquatiques**, d’**algues** et de **{term}`cyanobactéries`** est favorisée. Ce **processus naturel** fait partie du cycle de vie d’un plan d’eau et se déroule normalement sur une période de **plusieurs milliers d'années**.

Dans le sud du Québec, les lacs datent de la fin de la dernière période glaciaire, soit il y a environ 10 000 ans. Géologiquement, ces lacs sont considérés comme étant « jeunes », car ils se situent au début de leur cycle d’évolution et, selon les seuls cycles naturels, devraient être encore oligotrophes. On constate qu’il n’en est pas le cas. En effet, l’eutrophisation causée par les **activités humaines** se fait de façon accélérée, parfois sur **quelques dizaines d’années**. La quantité d’éléments nutritifs qui parvient aux plans d’eau est augmentée de façon importante par diverses **sources de pollution et d’{term}`érosion`**. Les **conséquences** affectent alors l’équilibre des **{term}`écosystèmes <Écosystème>` aquatiques**, mais aussi la **pêche**, la **baignade**, les autres **activités récréatives** et l’**approvisionnement en eau potable**. De plus, les conséquences peuvent s’avérer **dangereuses dans le cas d’apparition de fleurs d’eau de cyanobactéries**, car elles produisent des {term}`toxines <Toxine>`. Des cellules de cyanobactéries peuvent entrer dans les {term}`usines de production <Usine de production>` d’eau potable et perturber le traitement engendrant des risques pour la santé humaine.

Un plan d’eau oligotrophe présente une grande :

* biodiversité
* quantité d’oxygène dissous
* clarté et profondeur de l’eau

À l’inverse, lorsqu’un plan d’eau se transforme en milieu eutrophe, il y a graduellement une hausse de :

* végétaux aquatiques
* éléments nutritifs
* température de l’eau

### Quelles sont les causes?
L’eutrophisation {term}`anthropique` débute par une **accumulation significative** en :

* nutriments (principalement phosphore et azote);
* {term}`sédiments <Sédiment>` (particules de sol).

Les **apports en nutriments** proviennent de la pollution du ruissellement urbain; du lessivage des engrais et pesticides agricoles; des rejets d’eaux usées domestiques (installations septiques) ou municipales (surverses); des rejets industriels; des produits domestiques phosphatés; et des engrais domestiques. La croissance des végétaux est alors stimulée.

Les apports en **{term}`sédiments <Sédiment>`** proviennent de l’{term}`érosion` des sols du {term}`bassin versant` provoqué par le déboisement des rives et des terrains, par l’agriculture, par les coupes forestières, etc. L’envasement du lac est alors accéléré. Ces {term}`sédiments <Sédiment>` sont également une source de nutriments, ce qui favorise la croissance des plantes aquatiques.

En général, les problèmes sont surtout la contamination bactériologique, l’envasement, les proliférations de macrophytes et d’algues, la perte d’oxygène liée à la décomposition de ces dernières et la disparition graduelle du plan d’eau en entier.

### Quelles sont les conséquences?
En fin de processus, les changements dans la biodiversité animale et végétale signifient que les **espèces les mieux adaptées** aux nouvelles conditions sont favorisées. De plus, avec la prolifération des {term}`cyanobactéries` et la perte d’oxygène dans les milieux aquatiques, il peut résulter des **mortalités importantes de poissons**, une **perte économique** pour la valeur des résidences des riverains et les activités de la pêche. Dans le cas de lacs avec une forte concentration en cyanobactéries, cela peut être **dangereux pour les animaux** s’aventurant dans l’eau. Enfin, les plans d’eau contaminés et eutrophiés **détériorent la qualité** de l’approvisionnement et de la production en **eau potable**.

### Qu’en est-il des changements climatiques?
Il s’avère que l’**augmentation de température contribue à la croissance d’algues**, alors que l’**augmentation des précipitations augmente l’apport en sédiments** dans les cours d’eau. Les changements climatiques peuvent aussi affecter plusieurs autres facteurs qui altèrent la qualité des cours d’eau, notamment la physico-chimie des milieux.
