# Production
## Consommation d'eau journalière à Montréal
### Saviez-vous que l’eau potable n’est pas inépuisable?
Les {term}`eaux souterraines` contiennent plus de 30 % de l'eau douce du **monde entier** alors que les {term}`eaux de surface` n'en contiennent que 0,3 %. Au **Canada**, 26 % de la population dépend de l'eau souterraine pour son approvisionnement et sa consommation en eau potable [@eauenvironnement]. Au **Québec**, cette proportion est à 25 % [@eauxsouterraines]. Plus de 75 % de la population du Québec, soit 6,3 millions de personnes, est approvisionnée par une eau potable provenant d'eau de surface (fleuve, rivières et lacs)[@eaupotable] et 45 % de la population puise son eau potable dans le fleuve Saint-Laurent [@stlaurent].

**L’eau directement potable n’existe plus dans la nature au Québec**, toutes les sources d’eau du territoire doivent recevoir un traitement minimum, car il y a contamination par les animaux, l'agriculture, les industries, les rejets d’eaux usées, etc. Ainsi, la **production d'eau potable est nécessaire**. La **consommation d'eau est grande au Québec**, nous pouvons agir pour avoir des comportements plus responsables à cet égard, notamment en réduisant le gaspillage et les fuites.

## Quelle eau potable choisir : du robinet, de source, minérale ou traitée?
En plus de l’**eau potable de bonne qualité** dont nous avons la chance de disposer **sur demande au robinet**, une panoplie d’**eaux embouteillées** est **en vente**.

### Mais, quelles sont les différences? Laquelle devrions-nous boire?
Afin de mieux démystifier le sujet, il est essentiel de porter attention aux différentes natures de l’eau. Voici quelques informations concernant des normes et les distinctions entre les eaux potables du robinet, de source, minérale et traitée.

**Eau du robinet** : eau provenant d’une source {term}`souterraine <Eaux souterraines>` ou de {term}`surface <Eaux de surface>`, traitée et distribuée par un réseau public municipal et, au Québec, réglementée par le Ministère de l’Environnement, de la Lutte contre les changements climatiques, de la Faune et des Parcs (MELCCFP) via le [Règlement sur la qualité de l’eau potable (RQEP)](https://www.legisquebec.gouv.qc.ca/fr/document/rc/Q-2,%20r.%2040). Ce règlement établit les normes de qualité de l’eau potable, la fréquence de contrôle et vise à améliorer la protection de la santé publique par la mise à niveau de 77 normes, incluant 21 {term}`substances inorganiques <Substance inorganique>` et 50 {term}`substances organiques <Substance organique>`.

```{figure} images/M2.2.1_bouteilles.jpg
:alt: Bouteilles d'eau
:width: 934px
:align: left
L'eau embouteillée
```

**Eau embouteillée** : eau de diverses provenances selon sa nature et supervisée, au Québec, par le Ministère de l'Agriculture, des Pêcheries et de l'Alimentation (MAPAQ) via le [Règlement sur les eaux embouteillées](https://www.legisquebec.gouv.qc.ca/fr/document/rc/p-29,%20r.%202) mais dont les normes varient également selon la nature de l’eau.

**Eau de source** : eau provenant d’une source {term}`souterraine <Eaux souterraines>`, {term}`bactériologiquement pure <Eau bactériologiquement pure>` et non contaminée, contenant des sels minéraux dissous dont la teneur ne dépasse pas 1000 mg/L et devant respecter certaines normes concernant la concentration de 22 substances.

**Eau minérale** : Idem à l’eau de source, mais sans limite de concentration pour les {term}`sels minéraux` et les autres substances. Cette eau ne devrait pas être bue de façon exclusive car l’accumulation de sels minéraux qui pourrait en découler peut conduire à des problèmes de santé. Consommée régulièrement et en grande quantité, l'eau minérale peut avoir des effets nocifs sur les reins et peut à terme entraîner la formation de pierres. Il est recommandé pour les enfants, les personnes âgées et celles qui souffrent d'hypertension, de problèmes rénaux ou cardiovasculaires de consulter un spécialiste de la santé avant d'en faire une consommation régulière.

**Eau traitée** : eau ayant subi un traitement destiné à la rendre potable, provenant généralement du réseau de traitement et de distribution public municipal, puis pouvant être retraitée par un procédé d’osmose inverse, de charbon actif, de désinfection aux ultraviolets ou d’ozonation.

* Eau traitée minéralisée : eau de source ou eau traitée à laquelle on a ajouté des {term}`sels minéraux`.
* Eau traitée déminéralisée : eau de source ou eau traitée dont la teneur en {term}`sels minéraux` a été rendue inférieure à 10 mg/L par distillation. Cette eau a un goût astringent, non agréable, dû à l’absence de minéraux et ne devrait pas être bue de façon exclusive car les sels minéraux sont nécessaires au bon fonctionnement du corps humain. En effet, les minéraux peuvent contribuer à la fabrication d’hémoglobine ou de l'hormone thyroïdienne, ils interviennent dans la contraction musculaire, ils protègent le corps contre les radicaux libres ou encore participent à l'édification et au renouvellement du squelette. Si le corps ne reçoit pas suffisamment de sels minéraux, il pourra rapidement se retrouver en situation de carence et engendrer des conséquences multiples. Cette eau est surtout utile pour certains appareils comme les fers à repasser, les humidificateurs, etc.

Habituellement, les meilleures eaux sont celles qui ont des caractéristiques similaires à notre salive et qui ne vont pas trop s’éloigner de sa concentration en {term}`sels minéraux`. Au Québec, l’eau du robinet respecte le RQEP et nous avons accès à une eau de très bonne qualité!

### Y a-t-il des enjeux de développement durable?
Le principal **enjeu environnemental** de l’eau embouteillée est bien sûr la **matière plastique utilisée**. L’eau embouteillée n’est pas assujettie à l’ensemble des normes de qualité du MELCCFP de l’eau du robinet et elle présente en plus le problème du plastique, issue de l’industrie pétrolière et génératrice de **gaz à effet de serre**. La **fabrication** des bouteilles en plastique, leur **transport** et leur **recyclage** nécessitent beaucoup d’énergie et l’utilisation de ressources énergétiques non renouvelables. **Les bouteilles de plastique polluent et contaminent l’air, l’eau ainsi que les {term}`écosystèmes <Écosystème>`**.

```{figure} images/M2.2.1_VonWong_turnofftheplastic1.jpg
:alt: Von Wong turn off the plastic
:width: 1200px
:align: left
Enjeux de l'eau embouteillée illustrés par une oeuvre de [Benjamin Von Wong](https://www.vonwong.com/)
```
L’**enjeu économique** de l’eau embouteillée est également considérable compte tenu que l’eau du robinet est accessible et gratuite au Québec. Une part importante du prix de l’eau embouteillée **correspond en effet au contenant plutôt qu’au contenu**! Les compagnies d'eau embouteillée paient peu pour la ressource, donc le montant payé par le consommateur est davantage pour défrayer celui du plastique que pour l’eau, sans compter qu’il ne tient pas compte du coût des impacts environnementaux créés.

Le **coût de l’eau embouteillée** en format économique varie de **0,50 à 1,00 \$/L** (2,00 à 4,00 \$ par contenant de 4 litres) et, en consigne avec des bidons, varie de **0,26 à 0,42 \$/L** (5,00 à 8,00 \$ par contenant de 18,9 litres). En comparaison, le **traitement de l’eau municipale à Montréal** est au coût collectif de **0,0029 \$/L** (2,90 \$ pour 1000 litres). Ceci inclut la production d’eau potable, le traitement des eaux usées, la distribution, les infrastructures et les rénovations [@morissette].

:::{important}
Collectivement, c’est donc beaucoup plus intéressant environnementalement et économiquement d’investir dans le traitement communautaire que d’utiliser de l’eau embouteillée!
:::

## Jar test
