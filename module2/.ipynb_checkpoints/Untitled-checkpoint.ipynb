
:::{figure}
:label: egouts
:align: center
:class: grid grid-cols-2 items-end gap-4

(figure-egouts)=
![Réseau égouts 1](images/M3.2.2_Reseauegout_Mtl_1.jpg)
![Réseau égouts 2](images/M3.2.2_Reseauegout_Mtl_2.jpg)

Le réseau d’égout de Montréal [@montreal_sd_egout]
:::

### Quelles sont les différences entre les réseaux?

Le **{term}`réseau d’égout séparatif`** a un **avantage** important comparativement au {term}`réseau unitaire` : il **sépare les eaux usées des eaux pluviales** en envoyant les premières dans le réseau sanitaire, puis en envoyant les secondes dans le réseau pluvial tel que présenté au tableau suivant.

Eaux collectées et transportées par les réseaux d’égouts unitaire, séparatif et pseudo-séparatif
:::{table} Area Comparisons (written in fancy HTML)
:label: tbl:areas-html

<table>
   <tr>
      <th rowspan="2">Projection</th>
      <th colspan="3" align="center">Area in square miles</th>
   </tr>
   <tr>
      <th align="right">Large Horizontal Area</th>
      <th align="right" style="background: -webkit-linear-gradient(20deg, #09009f, #E743D9); -webkit-background-clip: text; -webkit-text-fill-color: transparent;">Large Vertical Area</th>
      <th align="right">Smaller Square Area
      <th>
   </tr>
   <tr>
      <td>Albers Equal Area</td>
      <td align="right">7,498.7</td>
      <td align="right">10,847.3</td>
      <td align="right">35.8</td>
   </tr>
   <tr>
      <td>Web Mercator</td>
      <td align="right">13,410.0</td>
      <td align="right">18,271.4</td>
      <td align="right">63.0</td>
   </tr>
   <tr>
      <td>Difference</td>
      <td align="right" style="background-color: red;color: white">5,911.3</td>
      <td align="right">7,424.1</td>
      <td align="right">27.2</td>
   </tr>
   <tr>
      <td>
         <bold>Percent Difference</bold>
      </td>
      <td align="right" style="background-color: green;color: white">44%</td>
      <td align="right">41%</td>
      <td align="right">43%</td>
   </tr>
</table>
:::
À noter : les eaux usées industrielles sont prétraitées avant d’être acheminées au réseau; les eaux pluviales incluent la pluie et le ruissellement; les eaux pluviales des bâtiments comprennent les eaux des drains de fondation et de toit.

Le **{term}`réseau d’égout unitaire`** achemine toutes les eaux collectées, qu’elles soient des eaux usées ou des eaux de ruissellement vers la {term}`station d’épuration`. Ce type de réseau est souvent le plus **ancien** et le **moins coûteux**. Un **effet d’autonettoyage** du réseau est créé par l’addition des eaux pluviales. Les conduites et l’usine de traitement sont dimensionnées pour prendre les eaux usées et les eaux des pluies les plus fréquentes. Ce n’est pas possible de construire pour les pluies intenses et rares, cela serait trop gros et trop cher. Ainsi, lors de fortes pluies ou de la fonte des neiges, qui peuvent être accentuées par les changements climatiques, la capacité hydraulique des égouts et de l’usine est dépassée. Ceci engendre donc des {term}`déversements <Déversement>` d’eaux usées, soit le **rejet d’eaux usées à l’environnement**. Les déversements d’eaux usées peuvent être des {term}`débordements <Débordement>` d’eaux usées planifiés à l’aide d’ouvrages de surverse avant la station, des débordements non planifiés directement dans les rues ou encore des {term}`dérivations <Dérivation>` d’eaux usées partiellement traitées à la station dû au contournement d’une étape de traitement.

```{figure} images/M3.2.2_et_M3.4_Deversement.jpg
:alt: Image d'un déversement
:width: 845px
:align: center
```

Le **{term}`réseau d’égout séparatif`** divise les eaux usées et les eaux pluviales dans **deux réseaux distincts**. Ainsi, les eaux pluviales ne passent pas par la station d’épuration, ce qui permet d’avoir une **quantité d’eau plus constante**. Toutefois, en milieu urbain particulièrement, il est tout de même recommandé de traiter les eaux pluviales au cours de leur passage dans le réseau avant de les rejeter dans le milieu naturel.

Le **{term}`réseau d’égout pseudo-séparatif`** est, comme son nom l’indique, un **système mixte** comprenant d’une part, un réseau unitaire pour les eaux usées et les eaux pluviales des bâtiments publics ou privés existants, puis d’autre part, un réseau distinct pour les eaux pluviales des voies publiques. Cette **solution intermédiaire** est mise en place lors de la transformation d’un réseau unitaire en réseau séparatif.
