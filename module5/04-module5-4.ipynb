# Eau et énergie
## Enjeux de l’hydroélectricité

Le **Canada** figure parmi les plus grands **producteurs mondiaux d'énergie hydroélectrique** et environ la moitié de sa production annuelle est issue des barrages du **Québec**. Les {term}`barrages-réservoirs <Barrage-réservoir>` ont façonné considérablement le paysage hydrique québécois et permettent de produire une énergie renouvelable, l'hydroélectricité. Ils ont créé de vastes bassins d’eau en amont et asséché les berges en aval. Plusieurs **{term}`écosystèmes <Écosystème>`** et des **populations** entières ont dû, par le passé, **s’adapter** aux changements et doivent, aujourd’hui encore, composer avec la gestion des niveaux d’eau induits par ces barrages. Les aménagements hydrauliques sont d’une **importance sociale et économique** majeure. Au Québec, **[la Loi sur la sécurité des barrages](https://www.legisquebec.gouv.qc.ca/fr/document/lc/s-3.1.01)**, entrée en vigueur en 2002 suite aux inondations du Saguenay de 1996, oblige les propriétaires de ces ouvrages à en évaluer régulièrement la **sécurité**, notamment sous l'effet des séismes et des {term}`crues <Crue>`.

Le contexte des **changements climatiques** rend l’étude, la surveillance et l’adaptation des ouvrages hydrauliques d’autant plus importantes. Partout dans le monde, les cours d’eau semblent soit se déchaîner, soit disparaître. Les manifestations des dérèglements climatiques sont nombreuses : les {term}`embâcles <Embâcle>`, le verglas, les {term}`crues <Crue>` et les inondations, les canicules et les sécheresses, l’{term}`érosion` et les glissements de terrain. Les capacités hydrauliques et les infrastructures sont impactées.

Les fortes précipitations et crues peuvent donner lieu à un **potentiel d'hydroélectricité** et contribuer à réduire la dépendance aux énergies fossiles. De plus, dans les projets de barrages, la production d'hydroélectricité est fréquemment complémentaire à d'autres finalités telles que la régulation des débits, le **contrôle des inondations**, la pisciculture, la villégiature, la réserve d’eau et l’irrigation. Par exemple, la République démocratique du Congo compte sur l’hydroélectricité pour la quasi-totalité de sa production d’énergie. Les Pays-Bas, en grande partie situés au niveau de la mer ou en dessous, ou encore la ville de Venise en Italie, sujette aux marées destructrices, perçoivent les infrastructures de contrôle des inondations comme une nécessité.

Toutefois, les barrages sont des **outils controversés** dans la lutte contre les changement climatiques. Ils engendrent des **impacts négatifs** sur la biodiversité et les écosystèmes. L’habitat essentiel et les lieux de reproduction de plusieurs espèces de poissons sont menacés [@he_2021]. Ces ouvrages hydrauliques présentent des risques environnementaux pour les {term}`milieux humides <Milieu humide>`, déjà fragiles et déclinants, dont la constitution protectrice et la biodiversité alimentent la vie [@viviano_2022]. Les terrains inondés lors de la création de réservoirs émettent des gaz à effet de serre (GES) plus nocifs que ce que laissaient croire les évaluations initiales. Les barrages peuvent aggraver le risque d’inondation en modifiant la composition et la structure des lits de rivière situés en aval [@Ma2022]. Enfin, les barrages sont dévastateurs pour l'eau, la terre et les gens qui y vivent puisqu'ils segmentent, transforment et détériorent le territoire. Les peuples autochtones sont particulièrement touchés. Ils ont toujours compris et reconnu l'importance de l'eau comme fondement essentiel à la vie et aux {term}`services écosystémiques <Service écosystémique>`; les {term}`milieux hydriques <Milieu hydrique>` façonnent leur mode de vie et présentent une valeur culturelle et spirituelle [@cgedu_sansdate].

## Enjeux des mesures d’efficacité d’eau

L’utilisation plus efficace de l’eau et de l’énergie est certainement une mesure essentielle pour réduire l’impact des villes, mais ces mesures peuvent avoir des **impacts sur la santé humaine**. La réduction des débits d'eau par l’installations d’**équipements sanitaires économes en eau** ainsi que la baisse de l'utilisation d'eau chaude peuvent conduire à des enjeux concernant la qualité microbiologique de l’eau et l'éclosion de bactéries. Ces enjeux doivent être identifiés et étudiés afin de proposer des pistes de solutions.

## Enjeux des mesures d’efficacité d’énergie

Les tours de refroidissement permettent de **climatiser** des bâtiments et de **refroidir** des procédés industriels et énergétiques tels que des centrales électriques, des installations de combustion, des industries. Elles évacuent vers le milieu extérieur la chaleur issue de systèmes de refroidissement en faisant circuler de l'eau. L’eau est refroidie par ruissellement en contact avec l’air.

Toutefois, l’usage de ces équipements comporte des **enjeux** concernant la présence de particules (pollen, insectes, poussières) dans l’eau et la prolifération de bactéries. Il en résulte une grande **consommation d’eau** et une **efficacité réduite** du système. De plus, la demande de climatisation est croissante en contexte de réchauffement planétaire. La recherche se penche sur des solutions plus sobres et appropriées. Les travaux de la professeure Émilie Bédard permettent d’améliorer l’efficacité énergétique des systèmes de refroidissement au moyen d’une filtration en continu qui en réduit les impacts.
