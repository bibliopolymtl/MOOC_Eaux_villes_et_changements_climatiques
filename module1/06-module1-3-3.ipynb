# Villes modernes
## Empreinte de l'eau

Le rôle de l'eau est essentiel au développement des villes modernes. De nombreux objets et produits sont fabriqués ou transformés en ville, que ce soit pour la construction, les activités quotidiennes ou l'alimentation de la population. L’empreinte de l'eau d’un produit est le **volume total d’eau** douce **utilisé directement ou indirectement** pour générer le produit alimentaire ou industriel, dans **toutes les phases** de sa **fabrication** et de sa **transformation**. Ainsi, cette mesure permet d’évaluer l’impact de produits qui nécessitent une forte utilisation d’eau dans leur production. L'empreinte de l'eau prend en compte trois catégories d'eaux.

* Eau bleue : {term}`eau de surface <Eaux de surface>` et {term}`eau souterraine <Eaux souterraines>` nécessaire (consommée ou évaporée).
* Eau verte : eau de pluie nécessaire (consommée ou évaporée).
* Eaux grises : eau douce nécessaire pour diluer les eaux usées (polluées) générées par la fabrication, afin de maintenir la qualité de l'eau, conformément aux normes nationales et locales.

## Développement des villes modernes
Entre 1750 et 1950, la population urbaine d'Europe et d'Amérique du Nord est passée de 10 % à 52% de la population totale de ces territoires. La tendance mondiale s'accélère, l'Asie et l'Afrique emboitent le pas. Aujourd'hui, **l'urbanisation doit être repensée**. Par exemple, avec **l’augmentation du niveau de la mer** de 1 mètre prévue pour 2100, les villes vont devoir reconsidérer leur organisation avec et non contre cet élément. De plus, de nombreux autres risques ou impacts des {term}`phénomènes météorologiques extrêmes <Phénomène météorologique extrême>` tels que les **vagues de chaleur, les fortes pluies, la sécheresse et les incendies de forêt** connexes sont à prévoir.
